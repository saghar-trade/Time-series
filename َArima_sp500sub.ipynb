```python
def plot_result(model, fulldata, train, test):
    # Get ARIMA model parameters
    params = model.get_params()
    d = params['order'][1]  # differencing order

    # In-sample prediction for training data
    train_pred = model.predict_in_sample(start=d, end=-1)

    # Out-of-sample prediction for test data with confidence intervals
    test_pred, confint = model.predict(n_periods=len(test), return_conf_int=True)

    # Plot results
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.plot(fulldata.index, fulldata, label='data')                     # Original data
    ax.plot(train.index[d:], train_pred, label='fitted')                # Fitted values
    ax.plot(test.index, test_pred, label='forecast')                    # Forecast values
    ax.fill_between(test.index, confint[:, 0], confint[:, 1],           # Confidence interval
                    color='red', alpha=0.3)
    ax.legend()

# Call the plotting function
plot_result(model, goog, train, test)
```

```python
def plot_test(model, test):
    # Predict for the test set with confidence intervals
    test_pred, confint = model.predict(n_periods=Ntest, return_conf_int=True)

    # Create plot
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.plot(test.index, test, label='true')                       # Actual test data
    ax.plot(test.index, test_pred, label='forecast')              # Forecasted values
    ax.fill_between(test.index, confint[:, 0], confint[:, 1],     # Confidence interval
                    color='red', alpha=0.3)
    ax.legend()

# Run the function
plot_test(model, test)
```

```python
#forecast RMSE

def rmse(t,y):
  return np.sqrt(np.mean((t-y)**2))
```

```python
# Compare RMSE for ARIMA model vs Naive prediction
print("RMSE ARIMA:", rmse(model.predict(Ntest), test))
print("RMSE Naive:", rmse(train.iloc[-1], test))
```
