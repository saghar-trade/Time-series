```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt


```

```python
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/air_passengers.csv' , index_col='Month' , parse_dates=True)
df.tail()
```

```python
df.isna().sum()
```

```python
df.index.freq='MS'
```

```python
df['LogPassengers']=np.log(df['Passengers'])
```

```python
Ntest=12
train=df.iloc[:-Ntest]
test=df.iloc[-Ntest:]
```

```python
!pip install numpy==1.26.4
!pip install --upgrade pmdarima

```

```python
import pmdarima as pm
```

```python
model=pm.auto_arima(train['Passengers'],
                   trace=True,
                    suppress_warnings=True,seasonal=True,
                    m=12
                    )
```

```python
model.summary()
```

```python
# Forecast on the test data
test_pred, confint = model.predict(n_periods=Ntest, return_conf_int=True)

# Plot forecast and confidence interval for test data
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(test.index, test['Passengers'], label='data')
ax.plot(test.index, test_pred, label='forecast')
ax.fill_between(test.index,
                confint[:, 0], confint[:, 1],
                color='red', alpha=0.3)
ax.legend();

# In-sample prediction (fitted values) for the training data
train_pred = model.predict_in_sample(start=0, end=-1)

# Plot fitted values for training data and forecast for test data
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(df.index, df['Passengers'], label='data')
ax.plot(train.index, train_pred, label='fitted')
ax.plot(test.index, test_pred, label='forecast')
ax.fill_between(test.index,
                confint[:, 0], confint[:, 1],
                color='red', alpha=0.3)
ax.legend();
```

```python
# Fit auto ARIMA model on log-transformed Passengers data
logmodel = pm.auto_arima(train['LogPassengers'],
                         trace=True,
                         suppress_warnings=True,
                         seasonal=True, m=12)
```

```python
# Display model summary
logmodel.summary()
```

```python
# Forecast for the test set
test_pred_log, confint = logmodel.predict(n_periods=Ntest, return_conf_int=True)
```

```python
# Plot the forecast vs actual data (log scale)
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(test.index, test['LogPassengers'], label='data')       # actual log data
ax.plot(test.index, test_pred_log, label='forecast')           # predicted log data
ax.fill_between(test.index, confint[:, 0], confint[:, 1],      # confidence interval
                color='red', alpha=0.3)
ax.legend();
```

```python
train_pred_log = model.predict_in_sample(start=0, end=-1)
```

```python
test_pred_log, confint_log = logmodel.predict(n_periods=len(test),
                                              return_conf_int=True)



train_pred = np.exp(train_pred_log)
test_pred = np.exp(test_pred_log)
confint = np.exp(confint_log)

# 5) Plot with aligned indices
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(train.index, train['Passengers'], label='train data')
ax.plot(test.index, test['Passengers'], label='test data')
ax.plot(train.index, train_pred, label='fitted')
ax.plot(test.index, test_pred, label='forecast')
ax.fill_between(test.index, confint[:,0], confint[:,1], color='pink', alpha=0.3)
ax.legend()
plt.show()
```

```python
#forecast RMSE

def rmse(t,y):
  return np.sqrt(np.mean((t-y)**2))
```

```python
print("NonLogRmse: ", rmse(test['LogPassengers'], test_pred))
print("LogRmse: ", rmse(test['Passengers'], np.exp(test_pred_log)))
```

### #Non seasonal

```python
modelns = pm.auto_arima(train['LogPassengers'],
                         trace=True,
                         suppress_warnings=True,
                        max_p=12,max_q=2, max_order=14,
                        stepwise=True,
                         seasonal=False)
```

```python
modelns.summary()
```

```python
test_pred_logns, confint_log = modelns.predict(n_periods=Ntest, return_conf_int=True)
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(test.index, test['LogPassengers'], label='data')       # actual log data
ax.plot(test.index, test_pred_logns, label='forecast')           # predicted log data
ax.fill_between(test.index, confint[:, 0], confint[:, 1],      # confidence interval
                color='red', alpha=0.3)
ax.legend();
```

```python
train_pred_logns = modelns.predict_in_sample(start=1, end=-1)

```

```python
# Plot fitted values for training data and forecast for test data

train_pred = np.exp(train_pred_logns)
test_pred = np.exp(test_pred_logns)
confint = np.exp(confint_log)

# 5) Plot with aligned indices
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(train.index[1:], train['Passengers'].iloc[1:], label='train data')
ax.plot(test.index, test['Passengers'], label='test data')


ax.plot(train.index[1:], train_pred, label='fitted')
ax.plot(test.index, test_pred, label='forecast')
ax.fill_between(test.index, confint[:,0], confint[:,1], color='pink', alpha=0.3)
ax.legend()
plt.show()



```

```python
rmse(test['Passengers'], np.exp(test_pred_logns))
```
