```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_pacf , plot_acf
```

```python
x0=np.random.randn(1000)
```

```python
plt.plot(x0)
```

```python
fig ,ax=plt.subplots(figsize=(10,8))
plot_pacf(x0, ax=ax);
```

```python
x1=[0]
for i in range (1000):
  x=0.5*x1[-1]+0.1*np.random.randn()   #x1[-1]  the last value of series
  x1.append(x)
x1=np.array(x1)

plt.plot(x1);
```

```python
fig ,ax=plt.subplots(figsize=(10,8))
plot_pacf(x1, ax=ax);
```

```python
x2=[0]
for i in range (1000):
  x=-0.5*x2[-1]+0.1*np.random.randn()
  x2.append(x)
x2=np.array(x2)

plt.plot(x2);
```

```python
fig ,ax=plt.subplots(figsize=(10,8))
plot_pacf(x2, ax=ax);
```

```python
x5=[0,0,0,0,0]
for i in range (1000):
  x=-0.5*x2[-1]-0.3*x5[-2]-0.6*x5[-5]+0.1*np.random.randn()
  x5.append(x)
x5=np.array(x5)

plt.plot(x5);
```

```python
fig ,ax=plt.subplots(figsize=(10,8))
plot_pacf(x5, ax=ax);
```

```python
#iid

fig ,ax=plt.subplots(figsize=(10,8))
plot_pacf(np.random.randn(1000), ax=ax);
```

```python
errors=0.1*np.random.randn(1000)
ma1=[]
for i in range (1000):
  if i>1:

    x=0.5*errors[i-1]+errors[i]
  else:
    x=errors[i]
  ma1.append(x)
ma1=np.array(ma1)

plt.plot(ma1);
```

```python
fig ,ax=plt.subplots(figsize=(10,8))
plot_acf(ma1, ax=ax);
```

```python
errors=0.1*np.random.randn(1000)
ma2=[]
for i in range (1000):


  x=0.5*errors[i-1]-0.3*errors[i-2]+errors[i]

  ma2.append(x)
ma2=np.array(ma2)

plt.plot(ma2);
```

```python
fig ,ax=plt.subplots(figsize=(10,8))
plot_acf(ma2, ax=ax);
```
