```python
import numpy as np
import matplotlib.pyplot as plt


# تولید 1000 نمونه از نرمال استاندارد (میانگین=0، واریانس=1)
samples = np.random.randn(1000)

print(samples[:10])
time_index = np.arange(len(samples))

# رسم سری زمانی
plt.figure(figsize=(10, 5))
plt.plot(time_index, samples, label="Random Samples")
plt.xlabel("Time Step")
plt.ylabel("Value")
plt.title("Time Series from Standard Normal Distribution")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# رسم هیستوگرام
plt.figure(figsize=(8, 5))
plt.hist(samples, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Standard Normal Samples")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# تولید ترند خطی

trend = np.linspace(0, 10, 1000)  # از 0 تا 10 در 1000 نقطه
```

Excercise #**2**

```python
# ترکیب نویز با ترند
noise=samples
data_with_trend = noise + trend
```

```python
plt.figure(figsize=(10, 5))
plt.plot(data_with_trend, label="Trend + Noise", color='steelblue')
plt.plot(trend, label="Trend Line", color='red', linewidth=2)
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Time Series with Trend and Noise")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(range(1000),data_with_trend, label="Trend + Noise", color='steelblue')
plt.plot(range(1000), trend, label="Trend Line", color='red', linewidth=2)
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Scatter Plot of Trend + Noise")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
cum_sum_noise = np.cumsum(noise)

# پلات
plt.figure(figsize=(10, 5))
plt.plot(cum_sum_noise, label="Cumulative Sum of Noise", color='blue')
plt.xlabel("Time")
plt.ylabel("Cumulative Value")
plt.title("Cumulative Sum of Gaussian Noise")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

```python
import numpy as np

# تعداد نمونه‌ها
n = 1000

# میانگین هر بعد
mean = [0, 0, 0]

# ماتریس کوواریانس (باید symmetric و مثبت‌ معین باشه)
cov = [
    [1, 0.8, 0.3],
    [0.8, 1, 0.2],
    [0.3, 0.2, 1]
]

# تولید داده‌ها
np.random.seed(0)
samples = np.random.multivariate_normal(mean, cov, size=n)

print("شکل داده‌ها:", samples.shape)
print(samples[:5])
```

```python
import matplotlib.pyplot as plt

#  بعد 1 و 2
plt.figure(figsize=(6,6))
plt.scatter(samples[:, 0], samples[:, 1], alpha=0.5)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter Plot of Multivariate Normal (Dim 1 vs Dim 2)")
plt.grid(True)
plt.show()


import pandas as pd
import seaborn as sns

df_samples = pd.DataFrame(samples, columns=["Feature 1", "Feature 2", "Feature 3"])
sns.pairplot(df_samples, diag_kind="hist")
plt.suptitle("Pairwise Scatter Plots of Multivariate Normal", y=1.02)
plt.show()
```
