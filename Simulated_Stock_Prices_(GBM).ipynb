```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#number of time steps
T=1000
#initial price
p0=10
#drift
mu=0.001


#last log price
last_p=np.log(p0)

log_returns=np.zeros(T)

prices=np.zeros(T)

for t in range(T):
  r=0.01*np.random.randn()
  p= last_p+ mu+ r
  log_returns[t]=mu+r
  prices[t]=np.exp(p)
  last_p=p
plt.figure(figsize=(20,8))
plt.plot(prices)
plt.title('Simulated Stock Prices (GBM)', fontsize=16)
plt.xlabel('Time Step', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
